```python
import os
os.getenv("GEMINI_API_KEY")
```

```python
from google import genai
```

```python
client=genai.Client()
```

```python
def gemini_llm(content):
    response=client.models.generate_content(
        model="gemini-3-flash-preview",
        contents=content
    )
    return response.text
```

```python
content=input("Enter your prompt:")
gemini_llm(content)
```
